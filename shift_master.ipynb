```python
import pandas as pd
import requests
import os
import time
from pathlib import Path
import logging
from typing import Set, Tuple
import sys
def collect_combinations(year: int,ps=False) -> Set[Tuple[str, str]]:
    """Get already processed player-team combinations for a given year."""
    trail ='ps' if ps else ''
    year_dir = Path(f"data/{year}{trail}")
    processed = set()
    total=[]
    if year_dir.exists():
        for file in year_dir.glob("*.csv"):
            nba_id = file.stem
            try:
                df = pd.read_csv(file)
                df['nba_id']=nba_id
                df['year']=year
                if trail =='ps':
                    df['playoffs'] = True
                else:
                    df['playoffs'] = False
                total.append(df)
                
            except Exception as e:
                logging.error(f"Error reading file {file}: {e}")
    #logging.info(processed)
    return pd.concat(total)

all_data=[]

for year in range(2001, 2025):
    print(year)
    trail='ps'
    processed = collect_combinations(year,ps=True)
    processed.to_csv('years/'+str(year)+trail+'.csv',index=False)
    all_data.append(processed)


# Process regular season (2001-2025)
for year in range(2001, 2026):
    print(year)
    trail=''
    processed = collect_combinations(year)
    processed.to_csv('years/'+str(year)+trail+'.csv',index=False)
```
